```python
# Plotting distribution of ratings
reviews.Rating.value_counts().plot(kind='bar')
```

```python
# Wordcloud plotting function
def plot_wordcloud(column):
    wordlist = [word for words in column for word in words.split(' ')]

    fdist_words = FreqDist(wordlist)

    wc = WordCloud(width=1000, height=800, max_words=100).generate_from_frequencies(fdist_words)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# Distinguishing positive and negative reviews
positive_reviews = reviews[reviews.Rating > 3]
negative_reviews = reviews[reviews.Rating < 3]
```

```python
# Displaying the most frequent words from positive reviews
plot_wordcloud(positive_reviews.Review)
```

```python
# Displaying the most frequent words from negative reviews
plot_wordcloud(negative_reviews.Review)
```
